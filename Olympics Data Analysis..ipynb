# Introduction

This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016.

This dataset provides an opportunity to ask questions about how the Olympics have evolved over time, including questions about the participation and performance of women, different nations, and different sports and events.

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Load the dataset
athletes=pd.read_csv("D:/athlete_events.csv")
regions=pd.read_csv("D:/noc_regions.csv")
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#Joining the Dataset

df=athletes.merge(regions,how="left",on="NOC")
```

```python
#Column names consistent

df.rename(columns={"region":"Region","notes":"Notes"},inplace=True);
```

```python
#Gold medal athletes

goldMedal=df[(df.Medal=="Gold")]
goldMedal.head()
```

```python
#Take only the values that are different from NaN.
goldMedal=goldMedal[np.isfinite(goldMedal["Age"])]
```

```python
#Gold beyond the age of 60.

goldMedal["Name"][goldMedal["Age"]>60].count()
```

```python
sporting_event=goldMedal["Sport"][goldMedal["Age"]>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medal for athletes over 60 years of age")
```

```python
#Gold medals from each country

goldMedal.Region.value_counts().reset_index(name="Medal").head(10)
```

```python
totalGoldMedals=goldMedal.Region.value_counts().reset_index(name="Medal").head(10)
```

```python
g=sns.catplot(x="index",y="Medal",data=totalGoldMedals,height=10,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per country")
```
